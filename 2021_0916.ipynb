```python
import requests
from bs4 import BeautifulSoup
header={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36"}
param={"_csrf": "f88a42af-6d7a-4219-92ad-72cd7ed47117",
"rideDate": "2021/09/16",
"station": "3300-臺中"}

url="https://www.railway.gov.tw/tra-tip-web/tip/tip001/tip112/querybystation"
html=requests.post(url,data=param,headers=header)
qtime=input("請輸入查詢時間(HH:MM)")+":00"

soup=BeautifulSoup(html.text,"lxml")
target_1=soup.find("tbody")
target_1=target_1.findAll("tr")
for row in target_1:
    tds=row.findAll("td")
    if (len(tds)>1 ):
        carTime=tds[1].text+":00"
        if(carTime>qtime):
            print(tds[0].text.replace("\n",""))
            print(tds[1].text)
            print(tds[2].text)
            print(tds[3].text.replace("\n",""))
            print()
        
```

```python
import requests
import json
import pandas as pd

header={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36"}
param={"SearchType": "S",
"Lang": "TW",
"StartStation": "TaiZhong",
"EndStation": "TaiPei",
"OutWardSearchDate": "2021/09/16",
"OutWardSearchTime": "20:00",
"ReturnSearchDate": "2021/09/16",
"ReturnSearchTime": "20:00"}

url="https://www.thsrc.com.tw/TimeTable/Search"
html=requests.post(url,data=param,headers=header)

html.encoding="utf-8"
html=html.text
highWay=json.loads(html)
station=highWay["data"]["DepartureTable"]["TrainItem"]

number=list()
startTime=list()
endTime=list()
duration=list()
for row in station:
    #指定時間的方式
    goTime=row["DepartureTime"]+":00"
    qTime="20:00:00"
    ##
    if(goTime>=qTime):
        number.append(row["TrainNumber"])
        startTime.append(row["DepartureTime"])
        endTime.append(row["DestinationTime"])
        duration.append(row["Duration"])
    
th=pd.DataFrame({"車次":number,
                "出發時間":startTime,
                "到達時間":endTime,
                "旅行時間":duration})
th
```
