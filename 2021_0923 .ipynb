```python
import requests
from bs4 import BeautifulSoup

url="https://supertaste.tvbs.com.tw/food"

html=requests.get(url)
html.encoding="UTF-8"


soup=BeautifulSoup(html.text,"lxml")
soup=soup.find(id="combolistUl")
foods=soup.findAll("li")
for i in foods:
    if len(i.findAll("a"))>0:
        
        
        link=i.findAll("a")[0].get("href")
        imgs=i.findAll("img")[0].get("data-original")
        print("https://supertaste.tvbs.com.tw"+link)
        print(imgs)
        print(i.findAll("a")[0].text)
     
```

```python
from selenium import webdriver
import time 
from bs4 import BeautifulSoup

'''放chromedriver的路徑'''
driver=webdriver.Chrome("C:\chromedriver.exe")

'''等待'''
driver.implicitly_wait(3)

driver.get("https://supertaste.tvbs.com.tw/food")


for i in range(20):
    '''網頁下拉 '''
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(1)

'''抓資料'''
soup=BeautifulSoup(driver.page_source,"lxml")
soup=soup.find(id="combolistUl")
foods=soup.findAll("li")
for i in foods:
    if len(i.findAll("a"))>0:
        
        
        link=i.findAll("a")[0].get("href")
        imgs=i.findAll("img")[0].get("data-original")
        print("https://supertaste.tvbs.com.tw"+link)
        print(imgs)
        print(i.findAll("a")[0].text)
     
```
