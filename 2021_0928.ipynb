```python
url="https://www.kkday.com/zh-tw/product/ajax_get_top_products?areaCode=A01-001-00019&upLimit=10&showLmit=4&csrf_token_name=1c39afe3220b3e2b2ce586854efba373"
import requests
import json

data=requests.get(url)
data.encoding="UTF-8"
data=data.text
#json.loads
items=json.loads(data)
all_item=items["data"]
for row in all_item:
    name=row["name"]
    introduction=row["introduction"]
    img_url=row["img_url"]
    price=row["price"]
    
    print(name)
    print(introduction)
    print(img_url)
    print(price)
    print()
```

有些json的某處會有贅字\
刪掉贅字

```python
import requests
url="https://24h.pchome.com.tw/cdn/onsale/v4/20210928/cvr1/data.json&27213825"

data=requests.get(url)
data.encoding="UTF-8"
data=data.text
data=data.split("var")
data=data[1]
key=data.find("[{")
data=data[key:-2]
items=json.loads(data)
for row in items:
    goods=row["Nodes"]
    for j in goods:
        link=j["Link"]["Url"]
        price=j["Link"]["Text2"]
        Img=j["Img"]["Src"]
        title=j["Img"]["Title"]
        
        print("https:"+link)
        print(price)
        print("https://b.ecimg.tw"+Img)
        print(title)
```

直接利用"Googlebot"來騙過伺服器


```python
import requests
from bs4 import BeautifulSoup
url="https://shopee.tw/search?keyword=switch"

#在header中設定
header={"cookies":"Googlebot","user-agent":"Googlebot"}

html=requests.get(url,headers=header)

html.encoding="UTF-8"

soup=BeautifulSoup(html.text,"lxml")
soup=soup.find_all(class_="col-xs-2-4 shopee-search-item-result__item")
for i in soup:
    link=i.find("a").get("href")
    name=i.find(class_="_10Wbs- _5SSWfi UjjMrh")
    prices=i.find(class_="zp9xm9 xSxKlK _1heB4J")
    print(link)
    print(name.text)
    print(prices.text)
    print()
```
