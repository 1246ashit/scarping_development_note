```python
import requests
from bs4 import BeautifulSoup

html=requests.get("https://www.ptt.cc/bbs/Food/index.html")

soup=BeautifulSoup(html.text,"lxml")
soup=soup.findAll("div",class_="r-ent")
for line in soup:
    print(line.find("a").text)
    print("https://www.ptt.cc"+line.find("a").get("href"))
    data=line.find("div",class_="date")
    print(data.text)
    print()
```
