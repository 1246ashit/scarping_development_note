# json解析

```python
import requests
import  json
'''json 用來解析json檔'''
url="http://citypark.tainan.gov.tw/App/parking.ashx?verCode=5177E3481D&type=1&"

'''.text 把檔案變成text檔'''
data=requests.get(url).text
#print(data)

'''json 檔案傳入'''
bike=json.loads(data)

for i in bike:
    print("站名:",i["address"])
    print()

```

# xml解析

```python
import requests
'''xml.sax 用來解析xml'''
'''建立class'''
import xml.sax
class BikeHandler(xml.sax.ContentHandler):
    def __init__(self):
        self.station=""
        self.rent=""
        self.space=""

    def startElement(self, name, attrs):
        self.tag=name
    
    '''因為xml標籤有頭尾'''
    def endElement(self, name):
        if self.tag=="StationName":
            print("站名:",self.station)
        elif self.tag=="AvaliableBikeCount":
            print("可借:",self.rent)
        elif self.tag=="AvaliableSpaceCount":
            print("可停:",self.space)
            
    def characters(self, content):
        if self.tag=="StationName":
            self.station=content
        elif self.tag=="AvaliableBikeCount":
            self.rent=content
        elif self.tag=="AvaliableSpaceCount":
            self.space=content

'''此if 是用來保證該class不能被其他文檔啟用'''
if __name__ =="__main__":
    paser=xml.sax.make_parser()
    bike=BikeHandler()
    
    '''paser.setContentHandler() 裡放的是ContentHandler , 因為bike繼承了ContentHandler'''
    paser.setContentHandler(bike)
    url="http://tbike-data.tainan.gov.tw/Service/StationStatus/Xml"
    data=requests.get(url)
    
     '''怕檔案讀進來變亂碼，用UTF-8來重新編碼'''
    data.encoding="UTF-8"
    data=data.text
    fileName="bike.xml"
    with open(fileName,"w",encoding="utf-8") as fobj:
        fobj.write(data)
    paser.parse(fileName)
```

```python
'''抓屬性'''
import requests
import xml.sax
class BikeHandler(xml.sax.ContentHandler):
    def startElement(self,name,attrs):
        if name=="Route":
            '''attrs丟入屬姓名'''
            print("路線:",attrs["nameZh"])
            print("起終點:",attrs["ddesc"])
            
if __name__ =="__main__":
    paser=xml.sax.make_parser()
    bus=BikeHandler()
    '''paser.setContentHandler() 裡放的是ContentHandler , 因為bike繼承了ContentHandler'''
    paser.setContentHandler(bus)
    url="http://ibus.tbkc.gov.tw/xmlbus/StaticData/GetRoute.xml"
    data=requests.get(url)
    data.encoding="UTF-8"
    data=data.text
    fileName="bus.xml"
    with open(fileName,"w",encoding="utf-8") as fobj:
        fobj.write(data)
    paser.parse(fileName)
```
